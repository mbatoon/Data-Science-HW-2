#  Create a program that:

1. Get 4 student names and 3 test scores for each student via input. Scores are between 0 to 100 <br> <br>
2. Calculate the average score for each student <br> <br>
3. Store the data in two lists ( one list for names and one list for average score) <br> <br>
4. Also calculate the class average <br> <br>
5. Print formatted summary, use string formatting, use one decimal <br> <br>


## Use this Notebook

## Use Markdown to title each step



## Example of expected final output 

==== Class Summary ==== <br>
Name &emsp;&emsp; Score

John  &emsp; &emsp;                 78.5 <br>
Elis  &emsp; &emsp; &nbsp;          65.6 <br>
Ali   &emsp;  &emsp; &nbsp; &nbsp;  75   <br>
Mark  &emsp; &emsp;                 88.3     <br>

Class Average: 76.8 <br>
Total Number of Students: 4 <br>

```python
names = []
averages = []

for i in range(4):
    name = input(f"Enter name for student {i+1}: ")
    S1 = float(input("Homework: "))
    S2 = float(input("Classwork: "))
    S3 = float(input("Exams: "))
    avg = (S1 + S2 + S3) / 3
    names.append(name)
    averages.append(avg)
    class_total = sum(averages)
    class_avg = class_total / len(averages)

print("\n====== Class Summary ======")
print(f"{'Name':<20}{'Score':<20}")

for i in range(4):
    print(f"{names[i]:<20}{averages[i]:.2f}")

print("-" *23)
print(f"Class Average: {class_avg:.1f}")
print(f"Total Number of Students: {len(names)}")
```
