```python
import numpy as np

categories = ["HW", "Project", "Exam 1", "Exam 2", "Final Exam"]
weights = np.array([0.15, 0.15, 0.20, 0.20, 0.30])

names = []
grades_matrix = np.zeros((10, 5))

for i in range(10):
    name = input(f"Enter name for student {i+1}: ")
    names.append(name)
    
    print(f"--- Entering grades for {name} ---")
    for j in range(5):
        grades_matrix[i, j] = float(input(f" Enter score for {categories[j]}: "))
```

```python
def get_letter_grade(score):
    if score >= 90: return 'A'
    elif score >= 80: return 'B'
    elif score >= 70: return 'C'
    elif score >= 60: return 'D'
    else: return 'F'
final_grades = np.dot(grades_matrix, weights)
```

```python
final_scores = np.dot(grades_matrix, weights)
letter_grades = [get_letter_grade(score) for score in final_scores]
```

```python
print("\n--- Final Student Grades ---")
print("=" * 45)
print(f"{'Student Name':<15} | {'Score':<8} | {'Grade'}")
print("-" * 45)

for i in range(10):
    print(f"{names[i]:<15} | {final_scores[i]:>8.2f} | {letter_grades[i]}")
```
